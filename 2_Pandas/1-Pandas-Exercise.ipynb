### Exploring Tortuosity

```python
##The following code was based off of ChatGPT using the following prompt and edited by Maia Griffith:
##"Can I find times of high vs low tortuosity to figure out when animals are spending more time at certain areas"
```

```python
# Function to calculate tortuosity
def calculate_tortuosity(trajectory):
    total_path_length = 0
    for i in range(1, len(trajectory)):
        # Need to use geodesic since lat/lon are not in meters
        total_path_length += geodesic(trajectory[i-1], trajectory[i]).meters

    # Find distance between first and last point
    straight_line_distance = geodesic(trajectory[0], trajectory[-1]).meters

    # Ensure straight_line_distance is not zero or NaN
    if straight_line_distance != 0:
        tortuosity = total_path_length / straight_line_distance
    else:
        tortuosity = np.nan  # Set tortuosity to NaN if the distance is invalid
        print("Invalid straight line distance")

    return tortuosity
```

```python
# Split data into hourly intervals and calculate tortuosity for each interval

# Calculate tortuosity for each group (i.e., each elephant's path)
tortuosity_values = []

for tag, group in dfClean.groupby('Tag'):
    # Sort the group by time (just in case the data is not sorted)
    group = group.sort_values(by='Datetime')

    # Iterate over the group and calculate tortuosity for time intervals (or other segmentation logic)
    for i in range(len(group) - 1):
        trajectory = np.array(list(zip(group['Longitude'][i:i+2], group['Latitude'][i:i+2])))  # 2-point interval for each segment
        tortuosity = calculate_tortuosity(trajectory)
        tortuosity_values.append({
            'Tag': tag,
            'Datetime': group['Datetime'].iloc[i],
            'tortuosity': tortuosity
        })
```

```python
# Plot tortuosity over time
plt.figure(figsize=(10, 5))
plt.plot(dfClean['Datetime'][1:], tortuosity_values, marker='o')
plt.title('Tortuosity Over Time')
plt.xlabel('Time')
plt.ylabel('Tortuosity')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Determine high and low tortuosity based on a threshold (e.g., mean tortuosity)
threshold = np.mean(tortuosity_values)  # For example, use the mean tortuosity as a threshold

# Classify high vs. low tortuosity
dfClean['tortuosity'] = pd.Series(tortuosity_values)
dfClean['high_tortuosity'] = dfClean['tortuosity'] > threshold
dfClean['low_tortuosity'] = dfClean['tortuosity'] <= threshold

# Print summary of periods of high and low tortuosity
high_tortuosity_periods = dfClean[dfClean['high_tortuosity']]
low_tortuosity_periods = dfClean[dfClean['low_tortuosity']]

print(f"High Tortuosity Periods:\n{high_tortuosity_periods[['Datetime', 'tortuosity']]}")
print(f"Low Tortuosity Periods:\n{low_tortuosity_periods[['Datetime', 'tortuosity']]}")
```

Now we are going to move on to something a bit more sophisticated: we are going to add a new field to our tracking data frame that represents the distance each record is from a given location. 

To do this, we first are going to create a **Python function** that computes the distance a supplied X and Y values is from a provided coordinate tuple. (_This doesn't work accurately with geographic coordinates, but we'll use it anyway..._)

```python
#Create a function that calculates the distance from a coordinate
def get_dist(theX, theY, thePoint):
    #Extract the coordinates from the provided point into separate variables
    center_x, center_y = thePoint
    #Get distances along X and Y axes
    x_dist = theX - center_x
    y_dist = theY - center_y
    #Compute the distances between points
    the_dist = (x_dist ** 2 + y_dist ** 2) * 0.5
    #Return the distance
    return the_dist
```

Next we need a point (just a tuple of X and Y coordinates) from which to measure distances. We'll use the centroid of our data (mean X and Y values).

```python
#Get the centroid of all track records
the_centroid = (df['X'].mean(),df['Y'].mean())
the_centroid
```

And here is the syntax for applying our `get_dist()` function to the X and Y series in our dataframe to compute each record's distance from `the_centroid`. 

```python
#Apply the function above to add a distance column
df['distance'] = get_dist(df['X'],df['Y'],the_centroid)
#Have a look
df.head()
```

Next we want to transform our data so that, for each hour of the day, we show the median distance each animal was recorded away from the overall centroid. 

To do this we want to transform or "pivot" our data such that:
* Each each unique`MovDataID` value becomes a row
* Each unique `Hour` value becomes a column
* The table shows mean distance values for each combination of elephant/hour of day

We'll do this and then add some styling such that dark blue values indicate animals furthest from the centroid.

```python
#Pivot data
df.pivot_table(
    index='MovDataID', # <- The index becomes the row name
    columns='Hour',    # <- Each unique value in the Hour column become a new separate column
    values=['distance'],      # <- The value assigned to each species/hour combination is taken from the "X" column
    aggfunc='mean'   # <- If more than one species/hour value exists, we will insert the median value
              ).style.background_gradient(cmap = 'YlGnBu') #Style the output
```
